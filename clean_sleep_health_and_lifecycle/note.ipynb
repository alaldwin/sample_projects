```python
df['Sleep Disorder'] = df['Sleep Disorder'].fillna('No Disorder')
```

```python
df.isnull().sum()
```

```python
df['Blood Pressure'].value_counts()
```

```python
bp = df['Blood Pressure'].astype(str).str.split("/", expand=True)

if bp.shape[1] == 2:
    df['Systolic'] = pd.to_numeric(bp[0], errors="coerce")
    df['Diastolic'] = pd.to_numeric(bp[1], errors="coerce")

df[['Blood Pressure','Systolic', 'Diastolic']]
```

```python
df['BMI Category'].value_counts()
```

```python
df["BMI Category"] = df["BMI Category"] \
    .astype(str) \
        .str.replace("Normal Weight", "Normal") \
            .str.strip()
df["BMI Category"].value_counts()
```

```python
df["Age"].describe()
```

```python
bins = [26, 34, 45, 60]

labels = ["27-34 (Early Career)", "35-45 (Mid Career)", "46-59 (Senior)"]

df["Age"] = pd.cut(df["Age"], bins=bins, labels=labels)

df["Age"].value_counts()
```

```python
df["Physical Activity Level"].describe()
```

```python
bins = [0, 45, 70, np.inf]

labels = ["Low", "Medium", "High"]

df["Physical Activity Level"] = pd.cut(df["Physical Activity Level"], bins=bins, labels=labels)

df["Physical Activity Level"].value_counts()
```
