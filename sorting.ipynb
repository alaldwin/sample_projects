```python
class Employee():
    def __init__(self, name, age, salary):
        self.name = name
        self.age = age
        self.salary = salary

    def __repr__(self):
        return '({}, {}, ${})'.format(self.name, self.age, self.salary)
    

e1 = Employee('carl', 37, 70000)
e2 = Employee('sarah', 29, 80000)
e3 = Employee('john', 43, 90000)

employees = [e1, e2, e3]

# custome function
# def e_sort(emp):
    # return emp.salary, name or age

# s_employee = sorted(employees) return a typeError: Unorderable types: Employee() < Employee()
s_employee = sorted(employees, key=lambda e: e.name, reverse=True)

print(s_employee)

```
