```python
# Write a function called is_prime, which checks if a number is prime.

def its_prime(n):

    if n <= 1:
        return False
    
    for i in range(2, int(n ** 0.5) + 1):
        if n % 2 == 0:
            return False
        
    return True

print(its_prime(7))
print(its_prime(2))
print(its_prime(4))
```

```python
# Create a function called show_args to take an arbitrary number of named arguments and print their names and values.

def show_args(**kwargs):

    result = []

    for key, value in kwargs.items():
        result.append(f"{key}: {value}")
    print("Received:", ", ".join(result))

print(show_args(name="Alice", age=30, city="New York"))
show_args(name="Bob", pet="Fluffy, the bunny")
```

```python
# Call your function factorial, it takes a whole number as a parameter and it return a factorial of the number  

def factorial(n):
    result = 1

    for i in range(1, n + 1):
        result *= i
    return result

print(factorial(5))
```

```python
def sum_of_odds(number):
    total = 0 

    for i in range(number + 1):
        if i % 2 == 1:
            total += i
    return total

print(sum_of_odds(10))
print(sum_of_odds(5))
```

```python
def even_numbers(number):

    total = 0 

    for i in range(number + 1):
        if i % 2 == 0:
            total += i
    return total

print(even_numbers(10))
print(even_numbers(5))
```

```python
def evens_and_odds(postive_integer):
    even = 0
    odd = 0

    for i in range(postive_integer + 1):
        if i % 2 == 0:
            even += i
        else:
            odd += i

    return f"The number of evens are {even}. The number of odds are {odd}."

    
print(evens_and_odds(100))
```

```python
def calculate_slope(y1, y2, x1, x2):
    if x2 - x1 == 0:
        return "Slope is undefined"
    
    m = (y2 - y1) / (x2 - x1)

    return m

print(calculate_slope(1, 2, 4, 6)) 
print(calculate_slope(2, 3, 2, 10))
```

```python
def solve_quadratic_eqn(ax, bx, c):
    if ax == 0:
        return "Not a quadratic equation"
    
    discriminant = bx ** 2 - 4 * ax * c

    if discriminant > 0:

        x1 = (-bx + discriminant ** 0.5) / (2 * ax)
        x2 = (-bx + discriminant ** 0.5) / (2 * ax)
        return x1, x2
    
    elif discriminant == 0:
        x = -bx / (2 * ax)
        return x,

    else:
        real_part = -bx / (2 * ax)
        imag_part = (abs(discriminant) ** 0.5) / (2 * ax)
        x1 = complex(real_part, imag_part)
        x2 = complex(real_part, -imag_part)
        return x1, x2


print(solve_quadratic_eqn(1, -3, 2))   
print(solve_quadratic_eqn(1, 2, 1))    
print(solve_quadratic_eqn(1, 1, 1))
```

```python
def reverse_list(array):
    reverse_arr = []

    for i in range(len(array)-1, -1, -1):
        reverse_arr.append(array[i])
    return reverse_arr

print(reverse_list(["A", "B", "C"]))
print(reverse_list([1, 2, 3, 4, 5]))
```

```python
# Write a lambda function which can solve a slope or y-intercept of linear functions.

# slope.
slope = lambda x1, x2, y1, y2: (y2 - y1) / (x2 - x1)
re = slope(2, 4, 6, 12)


# y-intercept
intercept = lambda x, y, m: y - m * x
inter = intercept(2, 12, 5)

print(re)
print(inter)
```
