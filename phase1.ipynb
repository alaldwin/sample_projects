```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\aldwin\OneDrive\Desktop\part 1\all_files\netflix_titles.csv")
df = df.copy()
df.head()
```

```python
print(df.shape)
print(df.columns)
df.isna().sum()
```

```python
df.director.unique()
```

```python
df = df.fillna({
    "director": "No Director",
    "cast": "No Cast",
    "country": "No Country",
    "date_added": "No date",
    "rating": "block"
})
df.head()
```

```python
df = df.drop_duplicates()
```

```python
df = df.drop("show_id", axis=1)
df.head()
```

```python
df.sort_values(by="release_year", ascending=False).reset_index()
df.head()
```

```python
plt.title("Movies Analysis")
plt.xlabel("Type")
plt.ylabel("Count")
df["type"].value_counts().plot(kind="bar")

plt.Figure(figsize=(12,4))
plt.show()
```

```python
plt.title("Movies Analysis")
plt.xlabel("Type")
plt.ylabel("Count")
df["director"].value_counts().head(10).plot(kind="bar")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.histplot(df["release_year"], kde=True)
plt.ylabel("Total Count")
plt.xlabel("years")
plt.title("Movies Year Analysis")
plt.Figure(figsize=(12,4))
```

```python
sns.scatterplot(data=df, x="release_year" , y="director")
```
