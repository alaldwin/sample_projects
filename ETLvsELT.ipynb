```python
# ETL
import pandas as pd

def extract(filename):
    return pd.read_csv(filename)

def transform(data_frame):
    return data_frame.loc[:, ["PassengerId", "Survived", "Name"]]

def load(data_frame, filename):
    data_frame.to_csv(filename)


extract_data = extract("titanic.csv").copy()

transform_data = transform(extract_data)

load_data = load(transform_data, "numbers_of_firms.csv")

print()
```
